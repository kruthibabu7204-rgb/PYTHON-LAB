```python
##union of two sets
a={1,2,3,4}
b={3,4,5,6}
union_set = a.union(b)
print("Union:",union_set)
```

```python
##isertion of two sets
a={1,2,3,4}
b={3,4,5,6}
intersection_set = a.intersection(b)
print("intersection:",intersection_set)
```

```python
##remove duplicates from a list using set
lst = [1,2,3,4,5]
unique_list = list(set(lst))
print("list without duplicates:",unique_list)
```

```python
##count frequency of characters using dictionary
string = "hello"
freq = {}
for char in string:
    if char in freq:
        freq[char] +=1
    else:
        freq[char] = 1
        print("character frequency:",freq)
```

```python
##create dictionary and search for a key
student = {
    "name": "kruthi",
    "age": 18,
    "course": "BEIS"
}
key = input("enter key to search: ")
if key in student:
    print("value:", student[key])
else:
    print("key not found")
```
