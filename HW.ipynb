```python
#creat a list of 5 number and print all elements
lst = [1, 2, 3, 4, 5]
print(lst)
```

```python
# print first and last elements of a list
lst = [10, 20, 30, 40, 50]
print("first element:", lst[0])
print("last element:", lst[-1])
```

```python
# add number 100 to the list
lst = [1, 2, 3, 4, 5]
lst.append(100)
print(lst)
```

```python
# remove element 3 from the list
lst = [1, 2, 3, 4, 5]
lst.remove(3)
print(lst)
```

```python
#find the sum of all elements
lst = [5, 10, 15, 20]
print("sum:",sum(lst))
```

```python
#find the maximum and minimum
lst = [8, 3, 1, 6]
print("Maximum:", max(lst))
print("Minimum:", min(lst))
```

```python
#print second and third element of tuple
t = (10, 20, 30, 40)
print("Second element:", t[1])
print("Third element:", t[2])
```

```python
#unpack tuple into variables and print
t = (5, 10, 15)
a, b, c = t
print(a, b, c)
```
