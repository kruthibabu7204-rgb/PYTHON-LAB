```python
#SETS(mutable,unordered,no duplicates)
numbers={1,2,3,2,1}
print("set:",numbers)
```

```python
#to adding number
numbers={1,2,3,5,6}
numbers.add(8)
numbers.add(9)
print("set:",numbers)
```

```python
#to discard the number
numbers={1,2,3,4,5,6,7}
numbers.discard(3)
numbers.discard(1)
print("set:",numbers)
```

```python
print("updated set:",numbers)
```

```python
#set operations
evens={2,4,6}
print("union:",numbers | evens)
```

```python
evens={2,4,6}
print("intersection:",numbers & evens)
```
